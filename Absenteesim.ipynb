# Train test split

```python
X = df.drop('Absenteeism time in hours',axis=1)
y = df['Absenteeism time in hours']
print("X shape ",X.shape)
print("y shape ",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,random_state=25, test_size =0.2)
print('X train shape', X_train.shape,'X test shape', X_test.shape)
print('y train shape', y_train.shape,'y test shape', y_test.shape)
```

```python
#Scaling
numeric_columns = []
for col in df.columns:
   if df[col].dtype == 'int64' or df[col].dtype == 'float64':
      numeric_columns.append(col)
```

```python
from sklearn.preprocessing import StandardScaler
ssc = StandardScaler()
df[numeric_columns] =ssc.fit_transform(df[numeric_columns])
```

# Feature Engineering

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info = mutual_info_classif(X_train, y_train, random_state = 20)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending = False)
```

```python
plt.figure(figsize = (20, 8))
mutual_info.sort_values(ascending = False).plot.bar();
```

```python
#train test split with important features
X = df.drop(['Absenteeism time in hours','Social smoker','Hit target','Pet','Height'],axis=1)
y = df['Absenteeism time in hours']
X_train, X_test, y_train, y_test = train_test_split(X, y ,random_state=25, test_size =0.3)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
Yhat=lm.predict(X_test)
```

```python
print("intercept_",lm.intercept_,"coef",lm.coef_)
```

```python
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error,explained_variance_score
r2_score(y_test, Yhat)
```

```python
lm.score(X_test,Yhat)
```

```python
mean_squared_error(y_test, Yhat)
```

```python
mean_absolute_error(y_test, Yhat)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pr = PolynomialFeatures(degree = 2,include_bias=False)
```

```python
X_train_poly = pr.fit_transform(X_train)
X_test_poly = pr.fit_transform(X_test)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train_poly, y_train)
```

```python
Yhat = lin_reg.predict(X_test_poly)
```

```python
mean_squared_error(y_test, Yhat)
```

```python
mean_absolute_error(y_test, Yhat)
```

```python
lin_reg.score(X_test_poly, Yhat)
```

# Ridge Regression

```python
Yhat = ridge_reg.predict(X_test)
```

```python
mean_absolute_error(y_test, Yhat)
```
